```python
def remove_outliers(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df

X = remove_outliers(X, selected_other_columns)

print(f"Shape after removing outliers: {X.shape}")
```

#### Model Evaluation

```python
def evaluate_pretrained_models(models, X_test, y_test):
    best_model = None
    best_pr_auc = 0
    results = {}    
    for model in models:
        print(f"Evaluating model: {model.__class__.__name__}")
        
        # Predict probabilities for ROC and PR curves
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
        
        # Classification Report
        print("\nClassification Report:")
        report = classification_report(y_test, y_pred, output_dict=True)
        print(classification_report(y_test, y_pred))
        
        # Confusion Matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='.2f', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
        plt.title(f"Normalized Confusion Matrix - {model.__class__.__name__}")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()
        
        # ROC-AUC
        if y_pred_proba is not None:
            fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
            roc_auc = auc(fpr, tpr)
            print(f"ROC-AUC Score: {roc_auc:.4f}")
            plt.figure()
            plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
            plt.plot([0, 1], [0, 1], 'k--', label="Random Chance")
            plt.title(f"ROC Curve - {model.__class__.__name__}")
            plt.xlabel("False Positive Rate")
            plt.ylabel("True Positive Rate")
            plt.legend(loc="lower right")
            plt.show()
        else:
            roc_auc = 0
            print("ROC-AUC not computed (model lacks `predict_proba`).")
        
        # PR-AUC
        if y_pred_proba is not None:
            precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
            pr_auc = average_precision_score(y_test, y_pred_proba)
            print(f"PR-AUC Score: {pr_auc:.4f}")
            plt.figure()
            plt.plot(recall, precision, label=f"PR Curve (AUC = {pr_auc:.2f})")
            plt.title(f"Precision-Recall Curve - {model.__class__.__name__}")
            plt.xlabel("Recall")
            plt.ylabel("Precision")
            plt.legend(loc="lower right")
            plt.show()
        else:
            pr_auc = 0
            print("PR-AUC not computed (model lacks `predict_proba`).")
        
        # Store results
        results[model] = {'roc_auc': roc_auc, 'pr_auc': pr_auc, 'classification_report': report}
        
        # Update best model based on PR-AUC (better for imbalanced data)
        if pr_auc > best_pr_auc:
            best_pr_auc = pr_auc
            best_model = model

    print("\nEvaluation complete.")
    print(f"Best model: {best_model.__class__.__name__} with PR-AUC = {best_pr_auc:.4f}")
    return best_model
```

```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_curve,
    auc,
    precision_recall_curve,
    average_precision_score
)
models=[]
models.append(best_lr_model)
models.append(best_svm_model)
models.append(best_rf_model)
models.append(best_ada_model)
models.append(best_xgb_model)
evaluate_pretrained_models(models, x_test, y_test)
```
