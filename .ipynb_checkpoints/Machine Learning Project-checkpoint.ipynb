# Importing the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
sns.set()
```

# Data Preprocessing

### Importing the Database

```python
raw_data=pd.read_csv('ml_datasource.csv')
raw_data.head()
```

```python
raw_data.describe()
```

```python
data=raw_data.copy()
```

### Removing Outliers

```python
fig, axes = plt.subplots(3, 2, figsize=(20,20))
sns.kdeplot(data=raw_data['days_on_platform'], ax=axes[0,0])
sns.kdeplot(data=raw_data['minutes_watched'], ax=axes[0,1])
sns.kdeplot(data=raw_data['courses_started'], ax=axes[1,0])
sns.kdeplot(data=raw_data['practice_exams_started'], ax=axes[1,1])
sns.kdeplot(data=raw_data['practice_exams_passed'], ax=axes[2,0])
sns.kdeplot(data=raw_data['minutes_spent_on_exams'], ax=axes[2,1])
```

```python
data_no_outliers = raw_data[raw_data['minutes_watched'] <= 1000]
data_no_outliers = data_no_outliers[data_no_outliers['courses_started'] <= 10]
data_no_outliers = data_no_outliers[data_no_outliers['practice_exams_started'] <= 10]
data_no_outliers = data_no_outliers[data_no_outliers['minutes_spent_on_exams'] <= 40]
```

```python
data_no_outliers.describe()
```

### Checking for Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
df=data_no_outliers.drop(['student_country'],axis=1)
vif_data["feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

print(vif_data)
```

```python
data_no_outliers=data_no_outliers.drop(['practice_exams_started'],axis=1)
data_no_outliers
```

```python
vif_data = pd.DataFrame()
df=data_no_outliers.drop(['student_country'],axis=1)
vif_data["feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

print(vif_data)
```

### Dealing with NaN Values

```python
data_no_outliers.loc[data_no_outliers['student_country'].isna()]
```

```python
data_no_outliers['student_country'] = data_no_outliers['student_country'].fillna("NAM")
data_no_outliers.loc[data_no_outliers['student_country'].isna()]
```

### Splitting the Data

```python
x=data_no_outliers.drop(['purchased'],axis=1)
y=data_no_outliers['purchased']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=365,stratify=y)
```

```python
x_test.shape
```

### Encoding the Data

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=170)
x_train['student_country_enc'] = enc.fit_transform(x_train['student_country'].to_numpy().reshape(-1, 1));
```

```python
x_test['student_country_enc'] = enc.transform(x_test['student_country'].to_numpy().reshape(-1, 1));
x_test['student_country_enc']
```

```python
x_test=x_test.drop(['student_country'],axis=1)
x_train=x_train.drop(['student_country'],axis=1)
x_test.head()
```

```python
x_train_array = np.asarray(x_train, dtype = 'float')
y_train_array = np.asarray(y_train, dtype = 'int')
x_test_array = np.asarray(x_test, dtype = 'float')
y_test_array = np.asarray(y_test, dtype = 'int')
```

# Creating a Logistic Regression Model

```python
import statsmodels.api as sm
log_reg=sm.Logit(y_train_array,x_train_array)
log_reg_results=log_reg.fit()
log_reg_results.summary()
```

```python
pred=log_reg_results.predict(x_test_array).round()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

3157 correctly predicted as 0 and 216 correctly predicted as 1
44 falsly predicted as 1 where they are actually 0 
109 falsly predicted as 0 where they are actually 1 

# Creating a K-Nearest Neighbors Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
from sklearn.model_selection import GridSearchCV
parameters_knn = {'n_neighbors':range(1, 51), 
              'weights':['uniform', 'distance']}
grid_search_knn = GridSearchCV(estimator = KNeighborsClassifier(), 
                           param_grid = parameters_knn, 
                           scoring = 'accuracy')
grid_search_knn.fit(x_train, y_train)
knn_clf=grid_search_knn.best_estimator_
```

```python
knn_clf
```

```python
grid_search_knn.best_estimator_
y_test_pred_knn=knn_clf.predict(x_test_array)
cm = confusion_matrix(y_test, y_test_pred_knn)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
print(classification_report(y_test, y_test_pred_knn))
```

# Creating a Support Vector Machines Model

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
scaler = MinMaxScaler(feature_range=(-1, 1))
x_train_array_scaled = scaler.fit_transform(x_train_array)
x_test_array_scaled = scaler.transform(x_test_array)
parameters_svc = {
    'kernel': ['linear', 'poly', 'rbf'],
    'C': range(1, 11),
    'gamma': ['scale', 'auto']
}
svc = SVC()
grid_search = GridSearchCV(estimator=svc, param_grid=parameters_svc, scoring='accuracy')
grid_search.fit(x_train_array_scaled, y_train)
best_svm_model = grid_search.best_estimator_
```

```python
best_svm_model
```

```python
y_test_pred_svc=best_svm_model.predict(x_test_array_scaled)
cm = confusion_matrix(y_test, y_test_pred_svc)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
print(classification_report(y_test, y_test_pred_svc))
```

# Creating a Decision Trees Model

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
dt_parameters = {
    'ccp_alpha': [0, 0.001, 0.002, 0.003, 0.004, 0.005]
}
dt = DecisionTreeClassifier(random_state=365)
grid_search = GridSearchCV(estimator=dt, param_grid=dt_parameters, scoring='accuracy')
grid_search.fit(x_train_array, y_train)
best_dt_model = grid_search.best_estimator_
```

```python
plt.figure(figsize=(15, 10))
plot_tree(best_dt_model,filled=True,feature_names=x_train.columns.values)
plt.show()
```

```python
y_test_pred_dt=best_dt_model.predict(x_test_array)
cm = confusion_matrix(y_test, y_test_pred_dt)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
print(classification_report(y_test, y_test_pred_dt))
```

# Creating a Random Forests Model

```python
best_dt_model
```

```python
rfc=RandomForestClassifier(ccp_alpha=0.001,random_state=365)
```

```python
rfc.fit(x_train_array,y_train_array)
```

```python
y_test_pred_rfc=rfc.predict(x_test_array)
cm = confusion_matrix(y_test, y_test_pred_rfc)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
print(classification_report(y_test, y_test_pred_rfc))
```
